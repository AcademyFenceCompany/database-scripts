```python
import pandas as pd
```

```python
original_customer_df = pd.read_csv("C:\\Users\\WebDev4\\Desktop\\afc_website\\db-scripts\\database-scripts\\customer_transformations\\customers\\customers_original.csv")
print(original_customer_df.head())
```

```python
customers_trans_df = pd.DataFrame(columns=['customer_id', 'original_customer_id', 'name', 'email', 'company', 'phone', 'phone_ext', 'alt_phone', 'alt_phone_ext', 'fax'])
```

```python
customers_trans_df['original_customer_id'] = original_customer_df['cust_id']
customers_trans_df['name'] = original_customer_df['name'].str.strip()
customers_trans_df['email'] = original_customer_df['email'].str.strip()
customers_trans_df['company'] = original_customer_df['company'].str.strip()
customers_trans_df['phone'] = original_customer_df['phone'].str.strip()
customers_trans_df['phone_ext'] = original_customer_df['ext'].str.strip()
customers_trans_df['alt_phone'] = original_customer_df['phone_alt'].str.strip()
customers_trans_df['fax'] = original_customer_df['fax'].str.strip()

customers_trans_df = customers_trans_df.fillna('NULL')
# print(customers_trans_df.tail(50))
```

```python
customers_trans_df.to_csv("C:\\Users\\WebDev4\\Desktop\\afc_website\\db-scripts\\database-scripts\\database_csvs\\customer_info\\customers_trans.csv", index=False)
```
